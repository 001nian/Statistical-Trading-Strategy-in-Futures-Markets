```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from datetime import datetime

data = pd.read_csv("./data/HO-5min.csv", parse_dates=[['Date', 'Time']])
data
```

```python
# Read the data
df = data

# Set the index to the "Date_Time" column
df.set_index("Date_Time", inplace=True)

# Plot the "Close" column
df["Close"].plot()

# Set the title and labels
plt.title("Price of HO vs Time (5 min intervals)")
plt.xlabel("Date_Time")
plt.ylabel("Close Price")

# Show the plot
plt.show()
```

```python
# Calculate the average price change per 5 mins
avg_price_change = df["Close"].diff().mean()

# Calculate the standard deviation of price
std_price = df["Close"].std()

# Calculate the average price of ticker
avg_price = df["Close"].mean()

# Calculate the max price and the date_time then
max_price = df["Close"].max()
max_price_date = df["Close"].idxmax()

# Calculate the min price and the date_time then
min_price = df["Close"].min()
min_price_date = df["Close"].idxmin()

# Calculate the max 5-min price increase
max_price_increase = df["Close"].diff().max()

# Calculate the min 5-min price decrease
min_price_decrease = df["Close"].diff().min()

# Print out the results
print("Average price change per 5 mins:", avg_price_change)
print("Standard deviation of price per 5 mins:", std_price)
print("Average price of ticker:", avg_price)
print("Max price:", max_price, "on", max_price_date)
print("Min price:", min_price, "on", min_price_date)
print("Max 5-min price increase:", max_price_increase)
print("Min 5-min price decrease:", min_price_decrease)
```

# Variance Ratio Test

## 1. Andrew Lo's Variance Ratio Test

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def variance_ratio(close_prices, k):
    close_prices = np.asarray(close_prices)
    one_volatility = np.std(np.diff(close_prices))
    k_volatility = np.std(close_prices[k:] - close_prices[:-k])
    return (k_volatility / (one_volatility * np.sqrt(k))) ** 2

file_path = "./data/HO-5min.csv"
data = pd.read_csv(file_path)
close_prices = data['Close']

lags = list(range(1, 10000))
ratios = [variance_ratio(close_prices, lag) for lag in lags]

plt.plot(lags, ratios, label='Variance Ratios')
plt.axhline(y=1, color='r', linestyle='--', label='Theoretical Value')
plt.xlabel('q values(minutes)')
plt.ylabel('Variance Ratio')
plt.title('Variance Ratio Test for HO (5-min) Futures')
plt.legend()
plt.show()
```

## 2. (Optional Extra) Chow Dennigs Variance Ratio Test

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from datetime import datetime

data = pd.read_csv("./data/COKE-5min.csv", parse_dates=[['Date', 'Time']]).dropna()
data
```

```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import scipy.stats as sps
```

```python
import warnings
warnings.filterwarnings("ignore")
max_t = 10000
prices = data["Close"]
returns = prices.pct_change()[1:]
n = len(returns)
variance_0 = returns.var()
variance_ratio_output = pd.DataFrame()
for t in np.arange(2, max_t+1):
    variance = returns.rolling(t).sum().var()
    variance_ratio = variance/(variance_0*t) - 1
    standard_error = (2*(2*t - 1)*(t-1)/(3*t*n))**(1/2)
    z_stat = variance_ratio/standard_error
    p_value = 2*(1 - sps.norm.cdf(abs(z_stat)))
    variance_ratio_output[t] = [round(variance_0, 6), round(variance, 6), round(variance_ratio, 4), round(standard_error, 4), round(z_stat, 4), p_value]
plt.plot(variance_ratio_output.iloc[4])
min_pvalue = min(variance_ratio_output.iloc[5])
max_z_stat = variance_ratio_output.iloc[4][variance_ratio_output.iloc[5] == min_pvalue].values[0]
Chow_Denning = 1 - (1 - min_pvalue)**(max_t)
if Chow_Denning > 0.1:
    print('the market is efficient')
elif max_z_stat > 0:
    print('the market is trending')
else:
    print('the market is mean-reverting')
```
